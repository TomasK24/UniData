```python
# but doing this one by one will also take some time so we can do multiple json files in one go   TOMAS KUKTA 21491719
from google.colab import files
import json
import pymongo
from bson import json_util
import re

# Step 1: Upload the JSON files
uploaded = files.upload()


# Step 2: Read the JSON files
data_dict = {
    'pastOrders': None,
    'ratings': None,
    'suppliers': None,
    'dailyInventoryRecord': None,
    'partners': None,
    'partnerHistory': None,
    'partners_old': None
}

# Function to normalize file names
def normalize_filename(filename):
    return re.sub(r'\s*\(\d+\)$', '', filename.split('.')[0])

for filename in uploaded.keys():
    with open(filename, 'r') as file:
        key = normalize_filename(filename)
        data = file.read()
        data_dict[key] = json_util.loads(data)

# Step 3: Connect to MongoDB and insert the data
client = pymongo.MongoClient("mongodb+srv://tkukta:X1W821BwJVdI3sGZ@uni.jzoumup.mongodb.net/?retryWrites=true&w=majority")
db = client['mydatabase']

collection_map = {
    'pastOrders': 'past_orders',
    'ratings': 'ratings',
    'suppliers': 'suppliers',
    'dailyInventoryRecord': 'daily_inventory_record',
    'partners': 'partners',
    'partnerHistory': 'partner_history',
    'partners_old': 'partners_old'
}

for key, data in data_dict.items():
    if data:
        collection_name = collection_map[key]
        collection = db[collection_name]
        try:
            result = collection.insert_many(data)
            print(f"Inserted {len(result.inserted_ids)} documents into {collection_name}")
        except pymongo.errors.BulkWriteError as e:
            print(f"Error inserting documents into {collection_name}: {e.details}")
```
