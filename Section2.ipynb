```python
#Tomas Kukta 21491719
# Importing the telecom dataset
url = 'https://raw.githubusercontent.com/Tomask24/UniData/main/telecom_dataset.csv'
telecom_df = pd.read_csv(url)
print(telecom_df.head())

# Load JSON files
customers_df = pd.read_json("customers.json")
daily_inventory_record_df = pd.read_json("dailyInventoryRecord.json")
partner_history_df = pd.read_json("partnerHistory.json")
partners_df = pd.read_json("partners.json")
partners_old_df = pd.read_json("partners_old.json")
past_orders_df = pd.read_json("pastOrders.json")
products_df = pd.read_json("products.json")
ratings_df = pd.read_json("ratings.json")
suppliers_df = pd.read_json("suppliers.json")

# Inspect data frames
print(customers_df.head())
print(past_orders_df.head())
print(products_df.head())
```

```python
#Tomas Kukta 21491719
# Merge customer-related data on Customer ID (_id)
merged_customers_df = pd.merge(customers_df, past_orders_df, left_on='_id', right_on='customer_id', how='left')

# Merge order-related data on Order ID (_id)
if 'order_id' in products_df.columns:
    merged_orders_df = pd.merge(past_orders_df, products_df, left_on='_id', right_on='order_id', how='left')
else:
    merged_orders_df = past_orders_df

# Combine all merged dataframes into one consolidated DataFrame
combined_df = pd.concat([merged_customers_df, merged_orders_df, partner_history_df, partners_df, partners_old_df, daily_inventory_record_df, ratings_df, suppliers_df], axis=0, ignore_index=True)
```

```python
#Tomas Kukta 21491719
# Convert dictionary and list columns to strings
for col in combined_df.columns:
    if combined_df[col].apply(lambda x: isinstance(x, (dict, list))).any():
        combined_df[col] = combined_df[col].apply(lambda x: str(x) if isinstance(x, (dict, list)) else x)

# Ensure data integrity by checking for duplicates and missing values
duplicate_rows = combined_df[combined_df.duplicated()]
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")
print("Missing values in each column:\n", combined_df.isnull().sum())

# Drop columns with more than 50% missing values
threshold = 0.5 * len(combined_df)
combined_df = combined_df.dropna(thresh=threshold, axis=1)

# Impute missing values for numerical columns with mean
for col in combined_df.select_dtypes(include=['float64']).columns:
    combined_df[col].fillna(combined_df[col].mean(), inplace=True)

# Impute missing values for categorical columns with mode
for col in combined_df.select_dtypes(include=['object']).columns:
    combined_df[col].fillna(combined_df[col].mode()[0], inplace=True)

# Verify the combined DataFrame
print(combined_df.head())
print(combined_df.info())
```

```python
import numpy as np

# Calculate common statistics
mean_values = combined_df.mean()
median_values = combined_df.median()
std_dev_values = combined_df.std()
min_values = combined_df.min()
max_values = combined_df.max()
quantiles = combined_df.quantile([0.25, 0.5, 0.75])

# Print statistics
print("Mean values:\n", mean_values)
print("Median values:\n", median_values)
print("Standard Deviation values:\n", std_dev_values)
print("Minimum values:\n", min_values)
print("Maximum values:\n", max_values)
print("Quantiles:\n", quantiles)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data

# Create a pair plot for numerical features
sns.pairplot(combined_df.select_dtypes(include=['float64']))
plt.show()

# Create a scatter plot for a pair of features, e.g., 'total day minutes' vs 'total day calls'
plt.scatter(combined_df['total day minutes'], combined_df['total day calls'])
plt.xlabel('Total Day Minutes')
plt.ylabel('Total Day Calls')
plt.title('Scatter Plot of Total Day Minutes vs Total Day Calls')
plt.show()

# Create a correlation matrix
corr_matrix = combined_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
